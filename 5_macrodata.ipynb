# Macro Collecte
L'objectif de ce notebook est de récupérer les données macroéconomiques qui nous intéresse pour en faire un fichier unique.

```python
import pandas as pd
```

```python
data = pd.read_sas("datas/construct_red.sas7bdat")
dates = data[data.CHRONIQUE == b"CHR2"].TRIMESTRE
```

```python
macros = pd.read_excel("datas/macro.xlsx")
```

```python
macros = macros.set_index(dates)
macros.drop("date",axis=1,inplace=True)
```
